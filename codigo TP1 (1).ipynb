```python
#cambiar region a codigo en la de 05

def regioncodigo (region): 
    if region == 'Gran Buenos Aires':
        return 1
    elif region == 'NOA':
        return 40
    elif region == 'NEA':
        return 41
    elif region == 'Cuyo':
        return 42
    elif region == 'Pampeana':
        return 43
    elif region == 'Patagónica':
        return 44
    
```
